```python
import pandas as pd

# Load the sales dataset
df= pd.read_csv(r"C:\Users\Harsh\Desktop\sales_data.csv")
```

```python
df
```

```python
# CREATE: Insert a new record using pd.concat
new_record = pd.DataFrame([{
    'Product_ID': 1100,
    'Sale_Date': '2025-06-29',
    'Sales_Rep': 'Alice',
    'Region': 'Central',
    'Sales_Amount': 5000.00,
    'Quantity_Sold': 25,
    'Product_Category': 'Electronics',
    'Unit_Cost': 200.00,
    'Unit_Price': 300.00,
    'Customer_Type': 'New',
    'Discount': 0.05,
    'Payment_Method': 'Online Transfer',
    'Sales_Channel': 'Online',
    'Region_and_Sales_Rep': 'Central-Alice'
}])

# Concatenate the new record to the existing DataFrame
df = pd.concat([df, new_record], ignore_index=True)
```

```python

### READ OPERATION ###
# Retrieve specific records: all sales made by 'Bob'
bob_sales = df[df['Sales_Rep'] == 'Bob']
print("Sales by Bob:")
print(bob_sales)
```

```python
### UPDATE OPERATION ###
# Increase discount to 0.10 for all 'New' customers in 'Electronics'
df.loc[(df['Customer_Type'] == 'New') & (df['Product_Category'] == 'Electronics'), 'Discount'] = 0.10
```

```python
### DELETE OPERATION ###
# Delete all records where 'Quantity_Sold' is less than 10
df = df[df['Quantity_Sold'] >= 10]
```

```python
### Save the updated DataFrame to a new CSV file
df.to_csv('updated_sales_data.csv', index=False)
print("\nCRUD operations completed. Updated dataset saved as 'updated_sales_data.csv'")
```
